## Decay Law and Other Functions
\begin{equation}
 A      (t)=A_{0}e^{-\lambda t}   
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt

Ao=50 # Initial Activity of radioactive substance in Curie
#Half Life of Ir192= 74 Years

t = np.linspace(0,74,10)
A = Ao*np.exp(-0.693*t/74)

#xerr = np.random.random_sample(10)
yerr = np.random.random_sample(10)

fig, ax = plt.subplots()

ax.errorbar(t, A,
            
            yerr=yerr,
            fmt='-o',capsize=5)


ax.set_xlabel('Time (Year)')
ax.set_ylabel('Activity (Curie)')
ax.set_title(' Decay Scheme')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
Ao=50 # Initial Activity of radioactive substance in Curie
#Half Life of Ir192= 74 Years
t = np.linspace(0,74,20)
A = Ao*np.exp(-0.693*t/74)
xerr = 5 #year
yerr = 5 # Curie

fig, ax = plt.subplots()

ax.errorbar(t, A,
            xerr=xerr,
            yerr=yerr,
            fmt='-o', color='blue',
             ecolor='black', elinewidth=2, capsize=2)

ax.set_xlabel('Time (Year)')
ax.set_ylabel('Activity (Curie)')
ax.set_title(' Decay Scheme')

plt.show()
```

## Plot Four functions plots in a single call to plot.

```python
import numpy as np
import matplotlib.pyplot as plt

# evenly sampled time at 200ms intervals
x = np.arange(1., 5., 0.4)

# red dashes, blue squares, green triangles and reverse dashes with opposite triangle
plt.plot(x, x, 'r--', x, x**2, 'bs', x, x**3, 'g^', x,1/x,'v--')

plt.show()
```

```python
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import numpy as np


x = np.linspace(0, 10, 1000)

fig, ax = plt.subplots()
ax.plot(x, np.sin(x), '-b', label='Sine')
ax.plot(x, np.cos(x), '--r', label='Cosine')
plt.ylabel('trigronometic function')
plt.xlabel('x(radian)')

ax.axis('equal')
leg = ax.legend();
```
